Plot the two following systems in python:

```python
import numpy as np
import math
from matplotlib import pyplot as plt
```

```python
#Make a linspace for each of the variables
lin_start = 0
lin_end = 2

t = np.linspace(lin_start, lin_end, 200)
A_0 = 10 #initial condition
k = 2
k_0 = 2
k_1 = 1

function1 = A_0 * math.e ** (-k * t)
function2 = (A_0-(k_0/k_1)) * math.e ** (-k * t) + (k_0/k_1)

figure1, axes = plt.subplots()
plt.plot(function1)

figure2, axes = plt.subplots()
plt.plot(function2)
```
